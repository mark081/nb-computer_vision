## Get rid of the Blue

```python
import cv2
```

```python
image = cv2.imread('Images/lenna.png')
```

```python
import matplotlib.pyplot as plt
```

```python
image.shape
```

```python
image[0,0,:] #First pixel value BGR
```

```python
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
rgb_image[0,0,:] #First pixel value RGB
```

```python
rgb_image[:,:,2] = 0
```

```python
plt.imshow(rgb_image)
```
